```python
import os
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
# Input data path
seis_path = '../data/seis/'
label_path = '../data/fault/'
```

```python
time_start= time.time()
seistrain = np.load(os.path.join(seis_path,'train/seis_train.npy'))
seisval = np.load(os.path.join(seis_path,'val/seis_val.npy'))
seistest = np.load(os.path.join(seis_path,'test/seis_test.npy'))

faulttrain = np.load(os.path.join(label_path,'train/fault_train.npy'))
faultval = np.load(os.path.join(label_path,'val/fault_val.npy'))
faulttest = np.load(os.path.join(label_path,'test/fault_test.npy'))

time_end = (time.time()-time_start)/60
print(f'Loading time used {time_end} mins')
```

```python
print('seis train shape: ',seistrain.shape)
print('fault train shape: ',faulttrain.shape)

print('seis val shape: ',seisval.shape)
print('fault val shape: ',faultval.shape)

print('seis test shape: ',seistest.shape)
print('fault test shape: ',faulttest.shape)
```

```python
def plot_random_images(images, labels, num_images=4):
    # Select random indices
    num_total_images = len(images)
    random_indices = np.random.choice(num_total_images, size=num_images, replace=False)
    
    fig, axes = plt.subplots(3, num_images, figsize=(18, 10))
    
    # Plot overlapped images
    for i, idx in enumerate(random_indices):
        axes[0, i].imshow(images[idx], cmap='seismic', aspect='auto')
        axes[0, i].axis('off')
        axes[0, i].set_title(f"seismic id: {idx}")
        axes[0, i].imshow(labels[idx], cmap='gray', aspect='auto',alpha=0.3)
        
    # Plot labels
    for i, idx in enumerate(random_indices):
        axes[1, i].imshow(labels[idx],cmap='gray', aspect='auto')
        axes[1, i].set_title(f"label id: {idx}")
        axes[1, i].axis('off')
    
    # Plot original images
    for i, idx in enumerate(random_indices):
        axes[2, i].imshow(images[idx], cmap='seismic', aspect='auto')
        axes[2, i].axis('off')
        axes[2, i].set_title(f"original seismic: {idx}")
        
    
    plt.tight_layout()
    plt.show()
```

```python
plot_random_images(seistrain,faulttrain,num_images=4)
```

```python
plot_random_images(seistrain,faulttrain,num_images=4)
```

```python
plot_random_images(seisval,faultval,num_images=4)
```

```python
plot_random_images(seisval,faultval,num_images=4)
```

```python
plot_random_images(seistest,faulttest,num_images=4)
```
